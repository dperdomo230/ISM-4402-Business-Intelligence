```python
import pandas as pd
import matplotlib.pyplot as plt
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,83,77,78,95]
GradeList = zip(names,grades)
df = pd.DataFrame(data = GradeList,
 columns=['Names', 'Grades'])
%matplotlib inline
df.plot()
# The text we want to display for our annotation 
displayText = "Wow!"
# xloc, yloc show the coordinates of the point we want to annotate 
xloc = 0.045
yloc = df['Grades'].min()
# xtext, ytext show  coordinates of where we want the text to appear using the coordinate system specified in textcoords
xtext = 8
ytext = -150
plt.annotate(displayText,
 xy=(xloc, yloc),
 arrowprops=dict(facecolor='black',
 shrink=0.05),
 xytext=(xtext,ytext),
 # xycoords sets the coordinate system to use to find the data point; it can be set separately for x and y            
 xycoords=('axes fraction', 'data'),
 textcoords='offset points')
```
